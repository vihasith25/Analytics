```python
import pandas as pd
import zipfile

# Path to your zip file
zip_path = "/content/companies data for analytics.zip"

# List to store DataFrames
data_frames = []

# Open the zip file and read each Excel file
with zipfile.ZipFile(zip_path, 'r') as z:
    for file in z.namelist():
        if file.endswith('.xlsx'):  # Check for Excel files
            with z.open(file) as f:
                data_frames.append(pd.read_excel(f))

# Concatenate all DataFrames into a single DataFrame
combined_data = pd.concat(data_frames, ignore_index=True)

# Shortlist companies by selecting the top 100 based on the "PAIDUP CAPITAL" and "Authorized Capital" columns
# Ensure "PAIDUP CAPITAL" and "Authorized Capital" are the correct column names in your dataset

'''Authorized Capital:
Best for assessing growth potential
   Paid-up Capital:
Best for assessing current financial strength
'''
a=int(input("enter no of companies: "))
shortlisted_data = combined_data.sort_values(by=["PAIDUP CAPITAL", "AUTHORIZED CAPITAL"], ascending=False).head(a)

# Save the shortlisted data to a CSV file
shortlisted_data.to_csv("top_100_companies_by_amount.csv", index=False)

# Optional: Display the shortlisted data
print("Top 100 companies by 'PAIDUP CAPITAL' and 'Authorized Capital':")
print(shortlisted_data)

# If you are using Google Colab and want to download the file:
from google.colab import files
files.download("top_100_companies.csv")

# Uncomment the following lines if you want to take a random sample of 100 companies
# sample_data = combined_data.sample(n=100, random_state=42)
# sample_data.to_csv("100_companies_data_for_analytics.csv", index=False)
```
